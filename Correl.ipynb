```python
import pandas as pd
import numpy as np

data = pd.read_csv('correl.csv')
data

data1 = data.loc[:, ["Calories", "Flavor"]]

#Create the Data Frame
data2 = pd.DataFrame(data1)
data2
```

```python
recipe=pd.DataFrame(data=data2)

from sklearn.impute import SimpleImputer
X = recipe.iloc[:,[0]].values
X = pd.DataFrame(data=X)
X
```

```python
imputer = SimpleImputer(missing_values = np.nan , strategy = 'mean')
imputer = imputer.fit(X)
X = imputer.transform(X) #it replaces missing data with mean
X = pd.DataFrame(data=X)
X

Y = recipe.iloc[:,[1]].values
Y = pd.DataFrame(data=Y)
Y

data3 = pd.concat([X, Y], axis=1, ignore_index=True)
data3
data3.columns =['Calories', 'Flavor']
```

```python
# f_oneway() function takes the group data as input and 
# returns F-statistic and P-value
from scipy.stats import f_oneway
 
# Running the one-way anova test between Flavor and Calories
# Assumption(H0) is that Flavor and Calories are NOT correlated
 
# Finds out the Prices data for each FuelType as a list
CategoryGroupLists=data3.groupby('Flavor')['Calories'].apply(list)
 
# Performing the ANOVA test
# We accept the Assumption(H0) only when P-Value &gt; 0.05
AnovaResults = f_oneway(*CategoryGroupLists)
print('P-Value for Anova is: ', AnovaResults[1])
```

```python
# P-value is below 0.05% hence Null hypothesis, 
# H0 = There is no correlation between Flavor and Calories, will get rejected. 
#And we will accept alternative hypothesis which is there is a correlation between Flavor and Calories
```
